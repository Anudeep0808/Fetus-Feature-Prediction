# Resizing

```python
from PIL import Image
import os

# Paths
input_directory = 'Test images'  # Directory containing the original images
output_directory = 'resized_test_images'  # Directory to save the resized images
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Image resizing
new_size = (224, 224)

for filename in os.listdir(input_directory):
    if filename.endswith(('.png', '.jpg', '.jpeg')):  # Add/check other formats if needed
        input_path = os.path.join(input_directory, filename)
        output_path = os.path.join(output_directory, filename)

        with Image.open(input_path) as img:
            resized_img = img.resize(new_size, Image.ANTIALIAS)
            resized_img.save(output_path)

print("Image resizing complete.")
```

# Contrast enhancement and noise reduction and normalization

```python
import cv2
import os

def preprocess_image(image_path, dest_path, clip_limit=4.0, tile_grid_size=(8, 8), median_ksize=5):
    # Load the ultrasound image in grayscale
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

    # Apply CLAHE for contrast enhancement
    clahe = cv2.createCLAHE(clipLimit=clip_limit, tileGridSize=tile_grid_size)
    enhanced_image = clahe.apply(image)

    # Apply a median filter for noise reduction
    denoised_image = cv2.medianBlur(enhanced_image, median_ksize)

    # Normalize the image
    normalized_image = cv2.normalize(denoised_image, None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_8U)

    # Save the preprocessed image
    cv2.imwrite(dest_path, normalized_image)

# Paths
source_directory = 'resized_test_images'
destination_directory = 'Test_images'

# Create the destination directory if it doesn't exist
if not os.path.exists(destination_directory):
    os.makedirs(destination_directory)

# Process each image in the source directory and save to the destination directory
for filename in os.listdir(source_directory):
    if filename.endswith('.png'):  # Check if the file is a PNG image
        file_path = os.path.join(source_directory, filename)
        dest_path = os.path.join(destination_directory, filename)
        preprocess_image(file_path, dest_path)
        print(f"Processed and saved: {filename}")
```
